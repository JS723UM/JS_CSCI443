```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Parameters for the standard Gaussian distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
#z = 2  # Z-value for the line and shaded area

# Generate x values for the Gaussian PDF
x = np.linspace(-4, 4, 1000)
gaussian_pdf = norm.pdf(x, mu, sigma)

# Plotting
plt.plot(x, gaussian_pdf, label='Gaussian PDF')

# Dotted line at Z=2 with label
plt.axvline(Z, color='k', linestyle=':')
plt.axvline(-Z, color='k', linestyle=':')

plt.text(Z+0.1, 0.34, f'Z={Z:.3f}', horizontalalignment='left')
plt.text(-Z-0.1, 0.34, f'-Z={-Z:.3f}', horizontalalignment='right')

# Shading the area where x < -Z and for x > Z
plt.fill_between(x, gaussian_pdf, where=(x > Z), color='orange', alpha=0.5)
plt.fill_between(x, gaussian_pdf, where=(x < -Z), color='orange', alpha=0.5)

# Adding labels and legend
plt.title('PDF of a Gaussian Distribution')
plt.xlabel('$D = \hat{p}_A - \hat{p}_B$')
plt.ylabel('Probability Density')
plt.legend()

plt.show()
```
