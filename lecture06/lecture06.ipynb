```python
### import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a synthetic dataset with 6 features
np.random.seed(0)  # For reproducibility
data = np.random.rand(100, 6)  # 100 samples, 6 features
columns = [
    'BMI',      # Body Mass Index
    'SysBP',    # Systolic Blood Pressure
    'DiasBP',   # Diastolic Blood Pressure
    'BloodSug', # Blood Sugar Level
    'Chol',     # Cholesterol Level
    'HRate',    # Heart Rate
]

# Create a DataFrame with shortened feature names
df = pd.DataFrame(data, columns=columns)

# Compute the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))  # Set figure size
sns.heatmap(corr_matrix, annot=True, fmt=".3f", cmap='coolwarm', 
            xticklabels=corr_matrix.columns, yticklabels=corr_matrix.columns)
plt.title('Correlation Matrix for Patient John Doe 1')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

# Output the correlation matrix as a mathematical matrix with 2 decimal places
print("Correlation Matrix:")
corr_matrix_string = corr_matrix.to_latex(float_format="{:0.3f}".format)
print(corr_matrix_string)
```
